```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('cleaned_shipments.csv')
```

```python
df.describe()
```

```python
# медиана задержек>0
df_delayed = df[df['delay_hours'] > 0]
overall_median = df_delayed['delay_hours'].median()
overall_median
```

```python
# медиана задержек по перевозчикам (median delay by carrier)
carrier_medians = df_delayed.groupby('carrier')['delay_hours'].median().reset_index()
carrier_medians
```

```python
carrier_medians = carrier_medians.sort_values(by='delay_hours').reset_index(drop=True)
carrier_medians.columns = ['Перевозчик', 'Медианное опоздание (ч)']
print("Медианное опоздание по каждому перевозчику:")
print(carrier_medians)                          
```
